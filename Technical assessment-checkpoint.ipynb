### Is the average number of listings per host significantly changing over time (year on year)?
#### Will use last review as time indicator however this will be a poor indicator as people can use an Air BnB and not leave a review. This can be very common also therefore even if some BnBs have reviews, that doesn't necessarily indicate the last person staying, it only indicates the last person to leave a review. 

#### Another method would be to use the calculated_host_listings_count and host_id collectively. Keep the rows which have the latest review per host_id, then take an average

```python
#check if there are any missing values in ids, host id, date
# run each line after each other
df['id'].isna().sum() #0
df['host_id'].isna().sum() #0
df['last_review'].isna().sum() #10052
```

```python
# create new dateframe and remove rows without date
df_listings = df[['calculated_host_listings_count','host_id','last_review','longitude','latitude','price']]
df_listings = df_listings.dropna(subset=['last_review'])
df_listings['date'] = pd.to_datetime(df_listings['last_review'], errors='coerce')
# make sure all dates are in the same format
df_listings['date'] = df_listings['date'].dt.strftime('%d/%m/%Y')
df_listings['year'] = df_listings['date'].str[-4:].astype(int)
df_listings
```

```python
# work out average of all listings
import statistics
statistics.mean(df_listings['calculated_host_listings_count'])
```

```python
# sort df_listings into year order then remove duplicated host_ids after seeing first host_id
df_listings = df_listings.sort_values(by="year")
df_listings.drop_duplicates(subset=['host_id'], keep = 'first')
```

```python
# for loop to work out average of listings per host_id each year and sums previous years into average of current year
year_array = []
sum_array = []
count_array = []


for year in range(2011,2020):
     
    sum_array.append(np.sum(df_listings.loc[(df_listings['year'] == year)].calculated_host_listings_count))
    count_array.append(df_listings.loc[df_listings['year'] == year].host_id.count())
    year_array.append(sum(sum_array)/sum(count_array))   
    
    print(year_array[-1])
    
plt.figure()
x = range(2011,2020)
plt.plot(x,year_array)
plt.xlabel('year')
plt.ylabel('average listings count')
    

```

```python
# split data into test, train, split
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
y = df_feature['price'].values
x = df_feature.loc[:,df_feature.columns!= 'price']
x = preprocessing.StandardScaler().fit(x).transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
#Training
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

Ks = 12
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
ConfustionMtx=[];
for n in tqdm(range(1,Ks)):
    neigh = KNeighborsClassifier(n_neighbors=n).fit(x_train, y_train)
    yhat = neigh.predict(x_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)
    std_acc[n-1] = np.std(yhat==y_test)/np.sqrt(yhat.shape[0])
mean_acc
```

```python
plt.plot(range(1,Ks),mean_acc)
plt.fill_between(range(1,Ks), mean_acc - (1 * std_acc), mean_acc + (1 * std_acc), alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors')
plt.tight_layout()
plt.show()

print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1) 

neigh = KNeighborsClassifier(n_neighbors=mean_acc.argmax()+1).fit(x_train, y_train)
```
